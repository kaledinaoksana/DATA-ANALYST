# Association Rules

## Confidence

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
rules
```

Вы предоставили данные, которые, похоже, представляют собой результаты анализа ассоциативных правил. Давайте разберем, что означает каждая колонка в этих данных:

- `antecedents`: Это наборы элементов или предшественников, которые появляются в ассоциативных правилах. Например, `(2_sausage)` означает набор элементов, включающий "сосиску" (2_sausage).

- `consequents`: Это наборы элементов или следствий, которые связаны с предшественниками в ассоциативных правилах. Например, `(1_frankfurter)` означает набор элементов, включающий "франкфуртер" (1_frankfurter).

- `antecedent support`: Доля транзакций или наблюдений, в которых встречаются предшественники.

- `consequent support`: Доля транзакций или наблюдений, в которых встречаются следствия.

- `support`: Доля транзакций или наблюдений, в которых одновременно встречаются и предшественники, и следствия. Это называется "поддержкой" для данного правила.

- `confidence`: Уровень уверенности в том, что следствия появятся, если уже есть предшественники. Рассчитывается как `support(предшественники и следствия) / support(предшественники)`.

- `lift`: Показатель, который показывает, насколько более вероятно появление следствия, если уже есть предшественники, по сравнению с их вероятностью появления независимо. Рассчитывается как `lift = support(предшественники и следствия) / (support(предшественники) * support(следствия))`.

- `leverage`: Измеряет разницу между фактической поддержкой правила и тем, что ожидалось, если предшественники и следствия были независимыми.

- `conviction`: Измеряет степень, с которой правило сильно зависит от предшественников.

- `zhangs_metric`: Метрика, используемая для оценки ассоциативных правил.

В приведенных вами данных показаны результаты нескольких ассоциативных правил. Каждое правило указывает, какие элементы (предшественники) связаны с другими элементами (следствиями) на основе поддержки, уверенности, лифта и других метрик.

## Lift

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```

## Lift and Confidence

```python
main = rules[(rules['lift'] >= 5) & (rules['confidence']>= 0.5)]
main
```
